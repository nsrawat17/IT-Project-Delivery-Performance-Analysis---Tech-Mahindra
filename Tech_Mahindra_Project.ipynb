# Q4. How does team size impact delivery?


```python
df.head()
```

```python
delayed = df[df['Status']=='Delayed']['Team_Size']
on_time = df[df['Status']=='Completed']['Team_Size']
```

```python
print("Average team size(Delayed) :",delayed.mean())
print("Average team size (on_time) :",on_time.mean())
```

# Larger teams may contribute slightly to better on-time delivery, but the impact is not very strong in this dataset.

```python
bins = [0,8,12, float('inf')]
labels =['Small', 'Medium','Large']

df['Team_Size_Category'] = pd.cut(df['Team_Size'], bins = bins , labels=labels)

print(df[['Team_Size','Team_Size_Category']].head())
```

```python
df.head(2)
```

```python
Avg_Delay_by_team_size =df.groupby('Team_Size_Category')['Delivery_Delay'].mean()
print("Avrage Delivery Delay by Team Size Category :\n",Avg_Delay_by_team_size)
```

```python
correlation = df[['Team_Size','Delivery_Delay']].corr()
```

```python
print("Correlation : \n" , correlation)
```

# Ans. Projects with larger teams  have lower average delivery delays.Average delay decreased from Small  Medium  Large teams, and a negative correlation of -0.004631 between team size and delivery delay confirms that larger teams are associated with faster delivery.
